```python
import pandas as pd
```

## Merging Cleaned Data

```python
# Read the csv file into a pandas DataFrame
eve_2014 = pd.read_csv('export/navidad_2014_12_24.csv')
after_2014 = pd.read_csv('export/navidad_2014_12_26.csv')

eve_2015 = pd.read_csv('export/navidad_2015_12_24.csv')
after_2015 = pd.read_csv('export/navidad_2015_12_26.csv')

eve_2016 = pd.read_csv('export/navidad_2016_12_24.csv')
after_2016 = pd.read_csv('export/navidad_2016_12_26.csv')

items = pd.read_csv('data/items.csv') # merge on 'item_nbr'

stores = pd.read_csv('data/stores.csv') # merge on 'store_nbr'

transactions = pd.read_csv('data/transactions.csv') # merge on 'date'
```

```python
# combined day data by year
all_2014 = eve_2014.append(after_2014, ignore_index=True)
all_2015 = eve_2015.append(after_2015, ignore_index=True)
all_2016 = eve_2016.append(after_2016, ignore_index=True)
```

```python
# combined all years
all_df = all_2014.append(all_2015, ignore_index=True)
all_df = all_df.append(all_2016, ignore_index=True)
all_df
all_df.to_csv('export/all_days.csv', index=False)
```

```python
# include relational data
all_df = all_df.merge(stores, how='inner')
all_df = all_df.merge(items, how='inner')
all_df = all_df.merge(transactions, how='inner')
all_df = all_df.sort_values(by=['date','store_nbr'])
all_df
all_df.to_csv('export/all_days_n_more.csv', index=False)
```
